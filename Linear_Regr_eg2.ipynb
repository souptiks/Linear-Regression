```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv('train.csv')
df
```

```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.xlabel('x',fontsize='10')
plt.ylabel('y',fontsize = '10')
plt.scatter(df.x,df.y,color = 'red',marker ='+')
plt.show()
```

```python
import numpy as np
reg = linear_model.LinearRegression()
input_data1 = np.array(df.x).reshape(-1,1)
input_data2 = np.array(df.x).reshape(-1,1)
reg.fit(input_data1,input_data2)
```

```python
df1 = pd.read_csv('test.csv')
```

```python
p = reg.predict(df1)
```
